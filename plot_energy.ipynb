```python
import numpy as np
import matplotlib.pyplot as plt

phi1 = np.linspace(-2*np.pi,2*np.pi,100)
phi2 = np.linspace(-2*np.pi,2*np.pi,100)
P1,P2 = np.meshgrid(phi1,phi2)

Ic = 50E-6
L= 5E-12
Ej = 2.067E-15*Ic/2/np.pi
Lj = 2.067E-15/Ic/2/np.pi

Ppos = (P1+P2)/2
Pneg = (P1-P2)/2

print(Lj/L)
print((2.067E-15/2/np.pi)**2/Ej/L)

E= (-np.cos(P1)-np.cos(P2)+Ppos**2/10)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(Ppos/np.pi,Pneg/np.pi,E)
plt.grid()
ax.clabel(CS, inline=1)
CB = fig.colorbar(CS)
```
